```python
plt.barh(df.Pclass.value_counts().index, df.Pclass.value_counts().values) # type: ignore
# plt.barh(df.Pclass.value_counts().index, df.Pclass.value_counts().values)
plt.xlabel('Pclass')
plt.ylabel("count of passenger")
plt.title('count per passenger in class')
plt.show()
```

```python
x=['Robert','Bolton','robb','jon-snow','Danerayas']
marks=[33,24,44,53,32]

plt.bar(x=x,height=marks,width=0.4,color=['orange','black'],edgecolor=['blue','red'],tick_label='start',ecolor='red')
plt.title('Marks obtained')
plt.ylabel('Makrs')
plt.xlabel('Names')
plt.show()
```

```python
fig, ax = plt.subplots()
x = ['Robb', 'Khal-Drago', 'jon-snow', 'bastard']
y = [40, 100, 30, 55]
bar_labels = ['shields', 'spear', 'swords', 'arrow']
bar_colors = ['tab:red', 'tab:blue', 'tab:pink', 'tab:orange']

ax.bar(x ,y, label=bar_labels, color=bar_colors)

ax.set_ylabel('Wars')
ax.legend(title='Weapons')
ax.set_title('Game of Thrones')
plt.show()
```

```python
x=[40,30,50,80]
y=['Robb', 'Khal-Drago', 'jon-snow', 'bastard']

plt.pie(x=x,labels=y, autopct='%1.1f%%')
plt.show()
```
