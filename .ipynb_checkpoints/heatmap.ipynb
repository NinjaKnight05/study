```python
import pandas as pd
import seaborn as sns
```

```python
df=sns.load_dataset('flights')
```

```python
df
```

```python
pivot=pd.pivot(data=df,values='passengers',index='year',columns='month')
pivot
```

```python
sns.heatmap(pivot,linewidths=0.1,linecolor='white',annot=True,fmt=".0f",)
```

```python
sns.heatmap(pivot,linewidths=0.1,linecolor='white',annot=True,fmt=".0f",robust=True,cmap="Blues")
```

```python
sns.heatmap(pivot,linewidths=0.1,linecolor='black',annot=True,fmt=".0f",robust=True,cmap="viridis" ,alpha=0.4)
```
