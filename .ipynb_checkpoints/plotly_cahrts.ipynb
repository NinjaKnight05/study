```python
import plotly.express as px
import pandas as pd
```

```python
df=px.data.iris()
df
```

```python
px.scatter(data_frame=df,x='sepal_length',y='petal_length',color='species',size='petal_width',hover_data='petal_width')
```

```python
px.scatter(data_frame=df,x='sepal_width',y='sepal_length',size='petal_width',color='petal_length')
```

```python
px.scatter(data_frame=df,x='sepal_width',y='sepal_length',size='petal_width',symbol='species',color='species')
```

```python

df["e"] = df["sepal_width"]/100
fig = px.scatter(df, x="sepal_width", y="sepal_length",color="species",error_x="e", error_y="e")
fig.show()
```
